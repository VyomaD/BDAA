**PDF - FacetGrid()**

```python
# Probability Density Function (PDF) based on symmetry_mean
sns.set_style('whitegrid')
sns.FacetGrid(df, hue='diagnosis', size=8) \
    .map(sns.distplot, 'symmetry_mean') \
    .add_legend()
plt.show()
```

Observations

1. The bars(Orange and Blue) are the histograms, and the curves represent the PDF based on Symmery_mean of the given breast cancer dataset.
2. There is a significant overlapping of data which amount to ambiguity
3. The values ranging from 0.10-16.5 are more diagnosed with benign tumor than the malignant cases
4. 0.15-0.20 are few in numbers who are diagnosed with malignant tumor


```python
# Probability Density Function (PDF) based on symmetry_mean
sns.set_style('whitegrid')
sns.FacetGrid(df, hue='diagnosis', size=8) \
    .map(sns.distplot, 'symmetry_se') \
    .add_legend()
plt.show()
```

Observations
1. Major overlapping can be seen in symmetry_se as compared to symmetry_mean and symmetry_worst.
2. As per the comparision the plot provides information about the number of malignant and benign diagnosis.

**Regression Plot**

How can we visualize different parameters linear relationships. This can be done with the help of regression model, regplot() and lmplot()

```python
sns.lmplot(data=df, x="concavity_worst", y="concavity_mean", hue="diagnosis", markers=["o", "x"], palette="Set1");
```

Here we can observe that by using lmplot(), we can combine regplot() and FacetGrid() to provide easy interface to show the linear regression on the 'faceted' plots that allows us to explore interactions with up to three additional categorical variables.

The best way to separate out a relationship is to plot both levels on the same axes and to use color to distinguish them. I have used markers to make plots reproduce better.

**Pair Plots**

Pair plot shows relationship between each column of the database. It also draws the univariate distribution plot of each variable on the diagnoal axis.

Pair plot : pairwise relationships in the dataset

```python
sns.pairplot(df, hue='diagnosis', vars=['radius_mean', 'radius_se', 'radius_worst'])
```

```python
sns.pairplot(df, hue='diagnosis', x_vars=['radius_mean', 'radius_se', 'radius_worst'], y_vars=['area_mean'], kind='reg')
```

```python
sns.pairplot(df, vars=["radius_worst", "perimeter_worst"], palette=sns.color_palette(['#FF1493', '#FF69B4']), hue='diagnosis', height=3)
```

Pairplot allows me to select specific columns in the database and will create plots. Comparision between radius_worst and perimeter_worst can be seen using pairplot, where average number diagnosed with malignant tumor are less in both matrices, and perimeter_worst contains more number of cases diagnosed with benign tumor.

**Joint Plot**

This plot allows me to plot two variables with bivariate and univariate graphs. We can perform meaningful conclusion and can also see trends and outliers.

```python
sns.jointplot(x="perimeter_mean", y="perimeter_worst", data=df, kind="reg");
```
